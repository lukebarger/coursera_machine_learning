Array max returns only the value not the index

```python
np.amax(A)
```

By default the index of the max is given for a flattened array.

```python
np.argmax(A)
```

unravel_index converts it back to an index for the given shape - arrayname.shape is the correct shape

```python
ind=np.unravel_index(np.argmax(A),A.shape)
ind
```

```python
A[ind]
```

Sum along specified axis.  Note 0 indexing is a thing

```python
M=np.random.random((8,8))
M
```

```python
np.sum(M,0)
```

```python
np.sum(M,1)
```

Eliminate all but diagonal elements of M by elementwise multiplication with identity

```python
M_diag=M*np.eye(M.shape[1])
M_diag
```

Sum along rows

```python
np.sum(M_diag,0)
```

Sum along rows then sum them all.  Here this is the sum of the diagonal elements of the original M array

```python
np.sum(np.sum(M_diag,0),0)
```
